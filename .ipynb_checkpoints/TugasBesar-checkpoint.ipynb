```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = 'mysecretpassword'
database = 'gdelt'
conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 1. View Event Data

```python
df = run("SELECT year, actiongeo_fullname, title, content   FROM gdelt_event where year in(2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017)")
df
```

### 2. Preposessing data - Mapping Province, Remove Nan/Null, Rename Column

```python
provinces = ['ACEH', 'SUMATERA UTARA', 'SUMATERA BARAT', 'RIAU', 'JAMBI', 'SUMATERA SELATAN', 'BENGKULU', 'LAMPUNG', 'KEPULAUAN BANGKA BELITUNG', 'KEPULAUAN RIAU', 'DKI JAKARTA' , 'JAWA BARAT', 'JAWA TENGAH', 'YOGYAKARTA', 'JAWA TIMUR', 'BANTEN', 'BALI', 'NUSA TENGGARA BARAT', 'NUSA TENGGARA TIMUR', 'KALIMANTAN BARAT', 'KALIMANTAN TENGAH', 'KALIMANTAN SELATAN', 'KALIMANTAN TIMUR', 'KALIMANTAN UTARA', 'SULAWESI UTARA', 'SULAWESI TENGAH', 'SULAWESI SELATAN', 'SULAWESI TENGGARA', 'GORONTALO', 'SULAWESI BARAT', 'MALUKU',  'MALUKU UTARA', 'PAPUA BARAT', 'PAPUA']

def parse_action_geo(v):
    arr_geo = v.split(',')
    return arr_geo[1] if len(arr_geo) > 2 else arr_geo[0]

clean_df = df.copy().drop_duplicates(subset = ['title'])
clean_df = clean_df[clean_df['title'].notnull()]
clean_df['province'] = clean_df['actiongeo_fullname'].apply(lambda v: parse_action_geo(v).strip().upper())
clean_df = clean_df.drop(columns = ['actiongeo_fullname'])
clean_df['province'] = clean_df['province'].str.replace('JAKARTA RAYA', 'DKI JAKARTA')
clean_df = clean_df[clean_df['province'].isin(provinces)]
clean_df = clean_df[['year', 'province', 'title', 'content']]
clean_df
```

### 3. Analytics - Ridge Regression 

```python
df_train = grouping_df.loc[grouping_df['year'].isin([2014, 2015, 2016])]
df_test  = grouping_df.loc[grouping_df['year'].isin([2017])]

X_train  = df_train[['province_encoded','ump',  'poverty']]
y_train  = df_train.crime

X_test  = df_test[['province_encoded','ump', 'poverty']]
y_test  = df_test.crime

from sklearn.linear_model import Ridge
rr = Ridge(alpha=0.01)

rr.fit(X_train, y_train)

rr_train_score = rr.score(X_train,y_train)
rr_test_score = rr.score(X_test, y_test)

print("ridge regression train score:", rr_train_score)
print("ridge regression test score :", rr_test_score)
print('Coefficients: \n', rr.coef_)

diabetes_y_pred = rr.predict(X_test)
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_test, diabetes_y_pred))

# grouping_df[['province_encoded','ump', 'population']]

# from sklearn import linear_model
# reg = linear_model.LinearRegression()

# reg.fit(x_train, y_train)
# #crime = m1*province_encoded + m2 * ump + m3 * population = b
# #score
# print(f"score  = {reg.score(x_train, y_train)}")
# #coef
# print(f"coef =  {reg.coef_}")
# #interception
# print(f"interception = {reg.intercept_}")
# # reg.predict([[ 3250000, 2250000]])

# from sklearn.linear_model import Ridge



# from sklearn import linear_model
# reg = linear_model.LinearRegression()

# reg.fit(x_train, y_train)
# #crime = m1*province_encoded + m2 * ump + m3 * population = b
# #score
# print(f"score  = {reg.score(x_train, y_train)}")
# #coef
# print(f"coef =  {reg.coef_}")
# #interception
# print(f"interception = {reg.intercept_}")
# # reg.predict([[ 3250000, 2250000]])

```
